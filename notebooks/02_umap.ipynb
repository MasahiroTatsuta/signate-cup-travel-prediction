```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score, accuracy_score
import seaborn as sns
import umap

train_data = pd.read_csv('../data/train_特徴量増加.csv')
test_data = pd.read_csv('../data/test_特徴量増加.csv')


train_data = train_data.drop("Unnamed: 0", axis=1)
test_data = test_data.drop("Unnamed: 0", axis=1)
# train_data = train_data.drop("Unnamed: 0.1", axis=1)
# test_data = test_data.drop("Unnamed: 0.1", axis=1)
train_data = train_data.drop("Age.1", axis=1)
test_data = test_data.drop("Age.1", axis=1)
# train_data = train_data.drop("CityTier.1", axis=1)
# test_data = test_data.drop("CityTier.1", axis=1)
# train_data = train_data.drop("NumberOfPersonVisiting.1", axis=1)
# test_data = test_data.drop("NumberOfPersonVisiting.1", axis=1)
# train_data = train_data.drop("NumberOfFollowups.1", axis=1)
# test_data = test_data.drop("NumberOfFollowups.1", axis=1)
# train_data = train_data.drop("PreferredPropertyStar.1", axis=1)
# test_data = test_data.drop("PreferredPropertyStar.1", axis=1)
# train_data = train_data.drop("NumberOfTrips.1", axis=1)
# test_data = test_data.drop("NumberOfTrips.1", axis=1)
train_data = train_data.drop("Passport.1", axis=1)
test_data = test_data.drop("Passport.1", axis=1)
# train_data = train_data.drop("PitchSatisfactionScore.1", axis=1)
# test_data = test_data.drop("PitchSatisfactionScore.1", axis=1)
train_data = train_data.drop("MonthlyIncome.1", axis=1)
test_data = test_data.drop("MonthlyIncome.1", axis=1)
train_data = train_data.drop("Marriage.1", axis=1)
test_data = test_data.drop("Marriage.1", axis=1)
# train_data = train_data.drop("Children.1", axis=1)
# test_data = test_data.drop("Children.1", axis=1)
train_data = train_data.drop("ProductPitched_Basic.1", axis=1)
test_data = test_data.drop("ProductPitched_Basic.1", axis=1)


train_data = train_data.drop("Children_1", axis=1)
test_data = test_data.drop("Children_1", axis=1)
train_data = train_data.drop("Marriage_1", axis=1)
test_data = test_data.drop("Marriage_1", axis=1)
train_data = train_data.drop("ProductPitched_King", axis=1)
test_data = test_data.drop("ProductPitched_King", axis=1)
train_data = train_data.drop("Occupation_Small_Business", axis=1)
test_data = test_data.drop("Occupation_Small_Business", axis=1)
train_data = train_data.drop("Occupation_Salaried", axis=1)
test_data = test_data.drop("Occupation_Salaried", axis=1)
# train_data = train_data.drop("Designation", axis=1)
# test_data = test_data.drop("Designation", axis=1)
# train_data = train_data.drop("TypeofContact", axis=1)
# test_data = test_data.drop("TypeofContact", axis=1)
train_data = train_data.drop("PreferredPropertyStar", axis=1)
test_data = test_data.drop("PreferredPropertyStar", axis=1)
train_data = train_data.drop("ProductPitched_Standard", axis=1)
test_data = test_data.drop("ProductPitched_Standard", axis=1)
train_data = train_data.drop("Occupation_Large_Business", axis=1)
test_data = test_data.drop("Occupation_Large_Business", axis=1)
train_data = train_data.drop("Agebin_41_45", axis=1)
test_data = test_data.drop("Agebin_41_45", axis=1)
train_data = train_data.drop("Agebin_34_40", axis=1)
test_data = test_data.drop("Agebin_34_40", axis=1)
train_data = train_data.drop("Agebin_31_33", axis=1)
test_data = test_data.drop("Agebin_31_33", axis=1)
train_data = train_data.drop("Agebin_28_30", axis=1)
test_data = test_data.drop("Agebin_28_30", axis=1)
train_data = train_data.drop("Agebin_15_27", axis=1)
test_data = test_data.drop("Agebin_15_27", axis=1)
# train_data = train_data.drop("NumberOfPersonVisiting", axis=1)
# test_data = test_data.drop("NumberOfPersonVisiting", axis=1)
train_data = train_data.drop("DurationOfPitch", axis=1)
test_data = test_data.drop("DurationOfPitch", axis=1)
train_data = train_data.drop("NumberOfFollowups_6", axis=1)
test_data = test_data.drop("NumberOfFollowups_6", axis=1)
# train_data = train_data.drop("ProductPitched_Super_Deluxe", axis=1)
# test_data = test_data.drop("ProductPitched_Super_Deluxe", axis=1)
# train_data = train_data.drop("Own_Car_未所持", axis=1)
# test_data = test_data.drop("Own_Car_未所持", axis=1)

train_data = train_data.drop("Abs(Marriage)/Marriage", axis=1)
test_data = test_data.drop("Abs(Marriage)/Marriage", axis=1)
train_data = train_data.drop("MonthlyIncome**3*exp(MonthlyIncome)", axis=1)
test_data = test_data.drop("MonthlyIncome**3*exp(MonthlyIncome)", axis=1)
train_data = train_data.drop("ProductPitched_Basic*Abs(Marriage)", axis=1)
test_data = test_data.drop("ProductPitched_Basic*Abs(Marriage)", axis=1)
train_data = train_data.drop("Age**2", axis=1)
test_data = test_data.drop("Age**2", axis=1)
train_data = train_data.drop("Age**2*ProductPitched_Basic", axis=1)
test_data = test_data.drop("Age**2*ProductPitched_Basic", axis=1)



umap_train_data = pd.read_csv('../data/train_特徴量増加.csv')
umap_test_data = pd.read_csv('../data/test_特徴量増加.csv')

train_data_np = umap_train_data.drop(columns=['id', 'ProdTaken']).to_numpy()
test_data_np = umap_test_data.drop(columns=['id']).to_numpy()

# TrainとTestを結合
combined_data_np = np.vstack((train_data_np, test_data_np))

# n_neighborsの候補
n_neighbors_list = [15, 20, 25, 30]


for i, n_neighbors in enumerate(n_neighbors_list):
    # DensMAPモデルの初期化
    densmap_model = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=0.1,
        n_components=2,
        random_state=42,
        densmap=True
    )

    # 結合データに対してDensMAPを適用
    combined_embedding = densmap_model.fit_transform(combined_data_np)

    # TrainとTestの埋め込みを分離
    train_embedding = combined_embedding[:len(train_data_np)]
    test_embedding = combined_embedding[len(train_data_np):]

    scaler = StandardScaler()
    train_embedding = scaler.fit_transform(train_embedding)
    test_embedding = scaler.transform(test_embedding)

    # DensMAP結果をPandas DataFrameに変換
    train_embedding_df = pd.DataFrame(train_embedding, columns=["DensMAP_1", "DensMAP_2"], index=train_data.index)
    test_embedding_df = pd.DataFrame(test_embedding, columns=["DensMAP_1", "DensMAP_2"], index=test_data.index)

    train_embedding_df2 = pd.concat([train_embedding_df, train_data[["id", "ProdTaken"]]], axis=1)
    test_embedding_df2 = pd.concat([test_embedding_df, test_data[["id"]]], axis=1)

    train_csv_path = f'../data/train_densmap_n{n_neighbors}.csv'
    test_csv_path = f'../data/test_densmap_n{n_neighbors}.csv'
    train_embedding_df2.to_csv(train_csv_path, index=False)
    test_embedding_df2.to_csv(test_csv_path, index=False)


    # 元のデータにDensMAP結果をconcat
    train_data_with_densmap = pd.concat([train_data, train_embedding_df], axis=1)
    test_data_with_densmap = pd.concat([test_data, test_embedding_df], axis=1)

    # 可視化（targetで色分けして不均衡を確認）
    plt.figure(figsize=(20, 12))
    sns.scatterplot(
        x="DensMAP_1",
        y="DensMAP_2",
        hue="ProdTaken",  # 実際のラベルで少数クラスを確認
        data=train_data_with_densmap,
        palette="deep",
        s=50,
        legend=False
    )
    plt.title(f"n_neighbors={n_neighbors}")
    plt.xlabel("DensMAP_1")
    plt.ylabel("DensMAP_2")
    plt.show()

    plt.figure(figsize=(20, 12))
    sns.scatterplot(
        x="DensMAP_1",
        y="DensMAP_2",
        data=test_data_with_densmap,
        palette="deep",
        s=100,
        marker="x",
        legend=False
    )
    plt.title(f"n_neighbors={n_neighbors}")
    plt.xlabel("DensMAP_1")
    plt.ylabel("DensMAP_2")
    plt.show()

    # CSVファイルに保存
    train_csv_path = f'../data/train_特徴量増加_densmap_n{n_neighbors}.csv'
    test_csv_path = f'../data/test_特徴量増加_densmap_n{n_neighbors}.csv'
    train_data_with_densmap.to_csv(train_csv_path, index=False)
    test_data_with_densmap.to_csv(test_csv_path, index=False)


# 結果の確認（最後のn_neighbors=50のデータ）
print("Train Data with DensMAP (n_neighbors=50):")
print(train_data_with_densmap.head())
print("\nTest Data with DensMAP (n_neighbors=50):")
print(test_data_with_densmap.head())








# # 可視化（オプション）
# plt.figure(figsize=(10, 6))
# sns.scatterplot(
#     x="UMAP_1",
#     y="UMAP_2",
#     data=train_data_with_umap,
#     palette="deep",
#     s=50,
#     label="Train",
#     legend="full"
# )
# plt.title("UMAP Embedding")
# plt.legend(title="Train")
# plt.show()

# plt.figure(figsize=(10, 6))
# sns.scatterplot(
#     x="UMAP_1",
#     y="UMAP_2",
#     data=test_data_with_umap,
#     palette="deep",
#     s=100,
#     marker="x",
#     label="Test",
#     legend="full"
# )
# plt.title("UMAP Embedding")
# plt.legend(title="Test")
# plt.show()





sns.set(style='darkgrid')
plt.figure(figsize=(50,12))
sns.heatmap(train_data_with_densmap.corr(), annot=True)

print(train_data.info())
```
